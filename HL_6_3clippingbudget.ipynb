```python
# Filename: experiment1_clip_9_to_1_enhanced_plot.py
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
import numpy as np
import matplotlib.pyplot as plt
from collections import OrderedDict
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, ConcatDataset, Subset

# Opacus
from opacus.validators import ModuleValidator
from opacus.grad_sample import GradSampleModule
from opacus.accountants import RDPAccountant

###############################################################################
# Recommended changes for ~20 epochs:
# (You can tweak these as desired)
###############################################################################
seed             = 0
batch_size       = 512
lr               = 0.05
outer_momentum   = 0.9
inner_momentum   = 0.10
noise_mult       = 1.0
delta            = 1e-5
num_epochs       = 20

# We'll do a smaller clip schedule from c_start=4.0 to c_end=2.0:
c_start          = 4.0
c_end            = 2.0

# We replicate the dataset 3 times for "self-augmentation"
self_aug_factor  = 3

# We'll keep a simpler schedule: decay at epoch 12, 18
SCHEDULE_MILESTONES = [12, 18]
SCHEDULE_GAMMA      = 0.1

M = 10000  # Max LRU momentum dictionary size

random.seed(seed)
torch.manual_seed(seed)
np.random.seed(seed)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

################################
# 1. Dataset
################################
mean, std = [0.4914, 0.4822, 0.4465], [0.2470, 0.2435, 0.2616]
train_tf = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.RandomErasing(p=0.5, scale=(0.02, 0.2)),
    transforms.Normalize(mean, std),
])
test_tf = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean, std),
])

train_ds = datasets.CIFAR10(root="./data", train=True, download=True, transform=train_tf)
test_ds  = datasets.CIFAR10(root="./data", train=False, download=True, transform=test_tf)
n = len(train_ds)


class IndexedSubset(Subset):
    def __getitem__(self, idx):
        x, y = super().__getitem__(idx)
        return x, y, idx

train_sub  = IndexedSubset(train_ds, range(n))
# replicate self_aug_factor times
from torch.utils.data import ConcatDataset
train_full = ConcatDataset([train_sub]*self_aug_factor)

train_loader = DataLoader(train_full, batch_size=batch_size, shuffle=True, drop_last=True)
test_loader  = DataLoader(test_ds,  batch_size=batch_size, shuffle=False)

##############################################
# 2. Enhanced ResNet20 with GroupNorm + ELU
#    and "GN after" the nonlinearity
##############################################
class BasicBlockEnhanced(nn.Module):
    """
    This block uses:
      - ELU instead of ReLU
      - GroupNorm AFTER the ELU
    """
    def __init__(self, in_planes, planes, stride=1):
        super().__init__()
        self.conv1 = nn.Conv2d(in_planes, planes, 3, stride, 1, bias=False)
        self.gn1   = nn.GroupNorm(8, planes)

        self.conv2 = nn.Conv2d(planes, planes, 3, 1, 1, bias=False)
        self.gn2   = nn.GroupNorm(8, planes)

        # Shortcut
        self.shortcut = nn.Sequential()
        if stride != 1 or in_planes != planes:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_planes, planes, 1, stride, bias=False),
                nn.GroupNorm(8, planes)
            )

    def forward(self, x):
        out = self.conv1(x)
        # ELU, then GroupNorm
        out = F.elu(out)
        out = self.gn1(out)

        out = self.conv2(out)
        out = F.elu(out)
        out = self.gn2(out)

        out = out.clone() + self.shortcut(x)
        return F.elu(out)

class ResNet20Enhanced(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()
        self.in_planes = 16
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1, bias=False)
        self.gn1   = nn.GroupNorm(8,16)

        self.layer1 = self._make_layer(16, 3, stride=1)
        self.layer2 = self._make_layer(32, 3, stride=2)
        self.layer3 = self._make_layer(64, 3, stride=2)

        self.avgpool= nn.AdaptiveAvgPool2d(1)
        self.fc     = nn.Linear(64, num_classes)

    def _make_layer(self, planes, num_blocks, stride):
        strides = [stride] + [1]*(num_blocks-1)
        layers = []
        for s in strides:
            layers.append(BasicBlockEnhanced(self.in_planes, planes, s))
            self.in_planes = planes
        return nn.Sequential(*layers)

    def forward(self, x):
        out = self.conv1(x)
        out = F.elu(out)
        out = self.gn1(out)

        out = self.layer1(out)
        out = self.layer2(out)
        out = self.layer3(out)

        out = self.avgpool(out)
        out = torch.flatten(out, 1)
        return self.fc(out)

def evaluate(model, loader):
    was_training = model.training
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for X, y in loader:
            X, y = X.to(device), y.to(device)
            preds = model(X).argmax(dim=1)
            correct += (preds==y).sum().item()
            total   += y.size(0)
    if was_training:
        model.train()
    return 100.*correct/total

################################
# 3. Build DP ResNet
################################
def build_model():
    net = ResNet20Enhanced().to(device)
    errs = ModuleValidator.validate(net, strict=False)
    if errs:
        net = ModuleValidator.fix(net).to(device)
    return GradSampleModule(net)

################################
# 4. Momentum Dict
################################
class LRUOrderedDict(OrderedDict):
    def __init__(self, *args, maxsize=10000, **kwargs):
        self.maxsize = maxsize
        super().__init__(*args, **kwargs)
    def __getitem__(self, key):
        val=super().__getitem__(key)
        self.move_to_end(key)
        return val
    def __setitem__(self, key, val):
        if key in self:
            self.move_to_end(key)
        super().__setitem__(key, val)
        if len(self) > self.maxsize:
            self.popitem(last=False)

momentum_dict = LRUOrderedDict(maxsize=M)

################################
# 5. gather per-sample grads
################################
def compute_per_id_momentum(dp_model, X, y, idxs):
    """
    1) forward+back => dp_model parameters have grad_sample
    2) gather them => group by ID => momentum
    Returns list of (id, grad) pre-clip
    """
    dp_model.zero_grad()
    out = dp_model(X)
    loss = F.cross_entropy(out, y)
    loss.backward()

    bs = X.size(0)
    param_vecs= [None]*bs

    for p in dp_model.parameters():
        gs = getattr(p, "grad_sample", None)
        if gs is None:
            continue
        gs_flat= gs.view(bs, -1).detach()
        for i in range(bs):
            if param_vecs[i] is None:
                param_vecs[i] = gs_flat[i]
            else:
                param_vecs[i] = torch.cat([param_vecs[i], gs_flat[i]], dim=0)
        p.grad_sample= None

    # group by real_id = idx%50000
    base_size= 50000
    group_dict= {}
    for i in range(bs):
        real_id= int(idxs[i].item()) % base_size
        v_i= param_vecs[i].to(device)
        if real_id not in group_dict:
            group_dict[real_id]= []
        group_dict[real_id].append(v_i)

    # apply momentum
    results= []
    for sample_id, v_list in group_dict.items():
        raw_v = torch.stack(v_list, dim=0).mean(dim=0)

        if sample_id in momentum_dict:
            old_v_half= momentum_dict[sample_id]
            old_v= old_v_half.to(device)
        else:
            old_v= torch.zeros_like(raw_v)

        # inner momentum
        new_v= (inner_momentum*old_v) + ((1.0 - inner_momentum)*raw_v)
        momentum_dict[sample_id]= new_v.half().cpu()

        results.append((sample_id, new_v))
    return results, len(group_dict)


################################
# 6. clip + sum + noise
################################
def clip_sum_noise(batch_v, clip_val, do_noise=True, return_postclip=False):
    """
    batch_v => list of (id, grad)
    clip each => sum => add noise => average => final_grad
    if return_postclip => also returns norms after clip
    """
    clipped_list= []
    postclip_norms= [] if return_postclip else None

    for (_, g) in batch_v:
        norm_g= g.norm(2)
        if norm_g> clip_val:
            g= g*(clip_val/(norm_g+1e-9))
        if return_postclip:
            postclip_norms.append(g.norm(2).item())
        clipped_list.append(g)

    grad_sum= torch.stack(clipped_list, dim=0).sum(dim=0)
    count= float(len(clipped_list))

    if do_noise:
        noise= torch.randn_like(grad_sum)*(noise_mult * clip_val)
        grad_sum += noise

    final_grad= grad_sum / count

    if return_postclip:
        return final_grad, postclip_norms
    else:
        return final_grad

def outer_step(dp_model, optimizer, final_grad):
    idx_start= 0
    for p in dp_model.parameters():
        numel= p.numel()
        chunk= final_grad[idx_start : idx_start+numel]
        p.grad= chunk.view_as(p)
        idx_start+= numel
    optimizer.step()

################################
# measure distribution => gather L1, L2, cos
################################
def measure_distribution(vecs):
    """
    given list of torch vectors
    returns (l1_list, l2_list, cos_list) vs. mean
    """
    if len(vecs)<1:
        return [],[],[]
    stacked= torch.stack(vecs, dim=0)
    mean_vec= stacked.mean(dim=0)
    mean_norm= mean_vec.norm(2).item()+1e-9

    l1_list= []
    l2_list= []
    cos_list=[]
    for v in vecs:
        l1_list.append(v.norm(p=1).item())
        l2_list.append(v.norm(p=2).item())
        dot_val= (v * mean_vec).sum().item()
        denom= v.norm(2).item()* mean_norm +1e-9
        cos_list.append(dot_val/ denom)

    return (l1_list, l2_list, cos_list)

def plot_hist(data, title, bins=50, color='blue'):
    """
    Simple helper => Displays figure on screen
    """
    plt.figure()
    plt.hist(data, bins=bins, alpha=0.7, color=color, edgecolor='black')
    plt.title(title)
    plt.xlabel("Value")
    plt.ylabel("Count")
    plt.show()


################################
# 7. main train
################################
def train_clip_budget_91():
    global momentum_dict
    momentum_dict.clear()

    dp_net= build_model().to(device)
    optimizer= optim.SGD(dp_net.parameters(),
                         lr=lr,
                         momentum=outer_momentum,
                         weight_decay=5e-4)

    # Optional LR scheduler
    scheduler= optim.lr_scheduler.MultiStepLR(
        optimizer,
        milestones=SCHEDULE_MILESTONES,
        gamma=SCHEDULE_GAMMA
    )
    accountant= RDPAccountant()

    total_steps= num_epochs * len(train_loader)
    step_count= 0

    for epoch in range(1, num_epochs+1):
        dp_net.train()
        losses= []

        # We track distribution from ALL batches in the epoch
        epoch_preclip_l1= []
        epoch_preclip_l2= []
        epoch_preclip_cos=[]
        epoch_postclip_l1=[]
        epoch_postclip_l2=[]
        epoch_postclip_cos=[]

        for i, (X,y,idxs) in enumerate(train_loader, start=1):
            step_count+=1
            frac= step_count/ float(total_steps)
            clip_val= c_start + (c_end- c_start)* frac

            X, y= X.to(device), y.to(device)

            # 1) gather momentum => batch_v
            batch_v, unique_count= compute_per_id_momentum(dp_net, X, y, idxs)

            # measure pre-clip distribution
            preclip_vecs= [g for (_, g) in batch_v]
            pre_l1, pre_l2, pre_cos= measure_distribution(preclip_vecs)
            epoch_preclip_l1.extend(pre_l1)
            epoch_preclip_l2.extend(pre_l2)
            epoch_preclip_cos.extend(pre_cos)

            # 2) clip => sum => noise => measure post-clip
            final_grad, postclip_norms= clip_sum_noise(
                batch_v, clip_val, do_noise=True, return_postclip=True
            )
            # replicate logic individually to measure full post-clip distribution
            clipped_vecs= []
            for (_, raw_v) in batch_v:
                nm= raw_v.norm(2).item()
                if nm> clip_val:
                    clipped_v= raw_v*(clip_val/(nm+1e-9))
                else:
                    clipped_v= raw_v
                clipped_vecs.append(clipped_v)

            post_l1, post_l2, post_cos= measure_distribution(clipped_vecs)
            epoch_postclip_l1.extend(post_l1)
            epoch_postclip_l2.extend(post_l2)
            epoch_postclip_cos.extend(post_cos)

            # 3) outer step
            outer_step(dp_net, optimizer, final_grad)

            # measure training loss
            with torch.no_grad():
                out= dp_net(X)
                loss_val= F.cross_entropy(out, y)
            losses.append(loss_val.item())

            # DP accountant step
            sample_rate= unique_count / 50000.0
            accountant.step(noise_multiplier=noise_mult, sample_rate=sample_rate)

        # End of epoch
        scheduler.step()
        acc= evaluate(dp_net, test_loader)
        eps= accountant.get_epsilon(delta)
        print(f"[Epoch {epoch:02d}] clip~{clip_val:.2f} "
              f"Loss={np.mean(losses):.3f} Acc={acc:.2f}% eps={eps:.2f}")

        # Plot distribution every 5 epochs (or you can do it every epoch if you want)
        if epoch % 5 == 0:
            # PRE-CLIP
            plot_hist(epoch_preclip_l1,
                      title=f"Pre-clip L1 norms (epoch={epoch})",
                      color='blue')
            plot_hist(epoch_preclip_l2,
                      title=f"Pre-clip L2 norms (epoch={epoch})",
                      color='green')
            plot_hist(epoch_preclip_cos,
                      title=f"Pre-clip cos sim vs. mean (epoch={epoch})",
                      color='red')
            # POST-CLIP
            plot_hist(epoch_postclip_l1,
                      title=f"Post-clip L1 norms (epoch={epoch})",
                      color='blue')
            plot_hist(epoch_postclip_l2,
                      title=f"Post-clip L2 norms (epoch={epoch})",
                      color='green')
            plot_hist(epoch_postclip_cos,
                      title=f"Post-clip cos sim vs. mean (epoch={epoch})",
                      color='red')

    final_acc= evaluate(dp_net, test_loader)
    final_eps= accountant.get_epsilon(delta)
    print(f"\nDone. Final test Acc={final_acc:.2f}% Eps={final_eps:.2f}")


################################
# 9. run
################################
if __name__=="__main__":
    train_clip_budget_91()
```
